```python
import matplotlib.pyplot as plt
import numpy as np

# 假设 RP 和 OD 的数据
RP = ['RP1', 'RP2', 'RP3', 'RP4']  # X轴 - RP 标签

# OD1 和 OD2 数据（两列数据和固定）
OD1 = np.array([10, 20, 15, 25])
OD2 = np.array([15, 10, 20, 15])

# 创建图形
fig, ax = plt.subplots()

# 绘制堆叠柱状图
ax.bar(RP, OD1, label='OD1', color='lightblue', edgecolor='black', hatch='/')  # 用斜线填充
ax.bar(RP, OD2, bottom=OD1, label='OD2', color='lightgrey', edgecolor='black', hatch='\\')  # 用反斜线填充

# 设置标签和标题
ax.set_xlabel('RP')
ax.set_ylabel('OD Value')
ax.set_title('RP vs OD Stacked Bar')

# 显示图例
ax.legend()

# 显示图形
plt.show()
```
